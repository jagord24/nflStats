```python
import pandas as pd
import seaborn as sns
```

```python
df = pd.read_csv('../data/working/NYG_pbp.csv')
```

```python
df = df[df['season'] == 2023]
```

```python
# create cumulative sum of play_differential column
df['play_differential'] = df['play_differential'].cumsum() 
```

```python
# plot play differential over season_seconds
# add vertical line at intervals of 3600
ax = sns.lineplot(x='season_seconds_elapsed', y='play_differential', data=df)
for i in range(1, max(df['week'])+1):
    ax.axvline(i * 3600, color='red', linewidth=0.3)
```
